## Cálculo Diferencial




Até agora, você aprendeu como avaliar limites para pontos em uma linha. Agora você vai desenvolver esse conhecimento e olhar para uma técnica de cálculo chamada * diferenciação *. Na diferenciação, usamos nosso conhecimento de limites para calcular a * derivada * de uma função para determinar a taxa de mudança em um ponto individual em uma linha.

Vamos nos lembrar do problema que estamos tentando resolver, aqui está uma função:

\begin{equation}f(x) = x^{2} + x\end{equation}

Podemos visualizar parte da linha que esta função define usando o seguinte código Python:

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Função
def f(x):
    return x**2 + x

# Cria uma matriz de valores x de 0 a 10 para o plot.
x = list(range(0, 11))

# Use a função para obter os valores y
y = [f(i) for i in x]

# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot 
plt.plot(x,y, color='blue')

plt.show()
```

Agora, sabemos que podemos calcular a taxa média de mudança para um determinado intervalo na linha calculando a inclinação de uma linha secante que conecta dois pontos na linha. Por exemplo, podemos calcular a mudança média para o intervalo entre x = 4 e x = 6 dividindo a mudança (ou * delta *, indicado como &Delta;) no valor de * f (x) * pela mudança no valor de * x *:


\begin{equation}m = \frac{\Delta{f(x)}}{\Delta{x}} \end{equation}


O delta para * f (x) * é calculado subtraindo os valores * f (x) * de nossos pontos, e o delta para * x * é calculado subtraindo os valores * x * de nossos pontos, assim:


\begin{equation}m = \frac{f(x)_{2} - f(x)_{1}}{x_{2} - x_{1}} \end{equation}


Então, para o intervalo entre x = 4 e x = 6, temos:

\begin{equation}m = \frac{f(6) - f(4)}{6 - 4} \end{equation}

Podemos calcular e plotar isso usando o seguinte código Python:

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Função
def f(x):
    return x**2 + x

# Cria uma matriz de valores x de 0 a 10 para o plot.
x = list(range(0, 11))

# Use a função para obter os valores y
y = [f(i) for i in x]

# Define os valores
x1 = 4
x2 = 6

# Obter os valores f (x) correspondentes
y1 = f(x1)
y2 = f(x2)

# Calcula a inclinação dividindo os deltas
a = (y2 - y1)/(x2 - x1)

# Cria uma matriz de valores x para a linha secante
sx = [x1,x2]

# Use a função para obter os valores y
sy = [f(i) for i in sx]

# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot da Função
plt.plot(x,y, color='blue')

# Plot dos intervalos
plt.scatter([x1,x2],[y1,y2], c='red')

# Plot da linha secante
plt.plot(sx,sy, color='magenta')

# Exibir a taxa média de mudança que foi calculada
plt.annotate('Mudança Média =' + str(a),(x2, (y2+y1)/2))

plt.show()
```

A taxa média de mudança para o intervalo entre x = 4 e x = 6 é <sup> 11 </sup> / <sub> 1 </sub> (ou simplesmente 11), significando que para cada ** 1 ** adicionado para * x *, * f(x) * aumenta em ** 11 **. 

Em outras palavras, se x representa o tempo em segundos e f(x) representa a distância em metros, a taxa média de mudança da distância ao longo do tempo (em outras palavras, * velocidade *) para o intervalo de 4 a 6 segundos é de 11 metros por segundo.

Até agora, isso é apenas álgebra básica; mas e se em vez da taxa média de mudança ao longo de um intervalo, queremos calcular a taxa de mudança em um único ponto, digamos, onde x = 5?

Uma abordagem que podemos usar é criar uma linha secante entre o ponto em que queremos o declive e outro ponto na linha de função que está infinitamente próximo a ela. Tão perto, na verdade, que a linha secante é, na verdade, uma tangente que passa por ambos os pontos. Podemos então calcular a inclinação para a linha secante como antes. Isso seria parecido com o gráfico produzido pelo seguinte código:

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Função
def f(x):
    return x**2 + x

# Cria uma matriz de valores x de 0 a 10 para o plot.
x = list(range(0, 11))

# Use a função para obter os valores y
y = [f(i) for i in x]

# Defina o ponto x1, arbitrariamente 5
x1 = 5
y1 = f(x1)

# Defina o ponto x2, muito próximo de x1
x2 = 5.000000001
y2 = f(x2)

# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot
plt.plot(x,y, color='blue')

# Plot do ponto
plt.scatter(x1,y1, c='red')
plt.annotate('x = ' + str(x1),(x1,y1), xytext=(x1-0.5, y1+3))

# Aproxima o declive da tangente e plote-o
m = (y2-y1)/(x2-x1)
xMin = x1 - 3
yMin = y1 - (3*m)
xMax = x1 + 3
yMax = y1 + (3*m)
plt.plot([xMin,xMax],[yMin,yMax], color='magenta')

plt.show()
```

## Calculando a Derivada

No código Python acima, criamos a linha secante (quase) tangencial especificando um ponto que está muito próximo do ponto em que queremos calcular a taxa de mudança. Isso é adequado para mostrar a linha conceitualmente no gráfico, mas não é uma maneira particularmente generalizável (ou precisa) de realmente calcular a linha para que possamos obter a taxa de alteração em qualquer ponto.

Se pelo menos soubéssemos uma maneira para calcular um ponto na linha o mais próximo possível de apontar com um determinado valor * x *.

Não, espere, nós já sabemos! É um limite :-).

Então, como podemos aplicar um limite neste cenário? Bem, vamos começar examinando nossa abordagem geral para calcular a inclinação um pouco mais detalhadamente. Nossa abordagem testada e comprovada é plotar uma linha secante entre dois pontos com valores diferentes de x, então vamos plotar uma linha arbitrária (* x, y *) e, em seguida, adicionar um valor arbitrário para * x *, que chamaremos de * h *. Então sabemos que podemos traçar uma linha secante entre (* x, f (x) *) e (* x + h, f (x + h) *) e encontrar sua inclinação.

Execute a célula abaixo para ver esses pontos:


```python
from matplotlib import pyplot as plt
%matplotlib inline

# Função
def f(x):
    return x**2 + x

# Cria uma matriz de valores x de 0 a 10 para o plot.
x = list(range(0, 11))

# Use a função para obter os valores y
y = [f(i) for i in x]

# Defina o ponto x1 e calcula y1
x1 = 3
y1 = f(x1)

# Define o incremento
h = 3

# Configura o ponto x + h
x2 = x1+h
y2 = f(x2)

# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot 
plt.plot(x,y, color='blue')

# Plot do ponto x, y
plt.scatter(x1,y1, c='red')
plt.annotate('(x,f(x))',(x1,y1), xytext=(x1-0.5, y1+3))

# Plot do ponto x + h
plt.scatter(x2,y2, c='red')
plt.annotate('(x+h, f(x+h))',(x2,y2), xytext=(x2+0.5, y2))

plt.show()
```

Como vimos anteriormente, nossa fórmula para calcular a inclinação é:

\begin{equation}m = \frac{\Delta{f(x)}}{\Delta{x}} \end{equation}

O delta para * f (x) * é calculado subtraindo os valores * f (x + h) * e * f (x) * dos nossos pontos, e o delta para * x * é apenas a diferença entre * x * e * x + h *; em outras palavras, * h *:

\begin{equation}m = \frac{f(x + h) - f(x)}{h} \end{equation}

O que realmente precisamos é da inclinação na menor distância possível entre x e x + h, então estamos procurando o menor valor possível de * h *. Em outras palavras, precisamos do limite quando * h * se aproxima 0.

\begin{equation}\lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \end{equation}

Esta equação é generalizável, e podemos usá-la como a definição de uma função para nos ajudar a encontrar a inclinação em qualquer valor dado de * x * na linha, e é o que chamamos de * derivada * de nossa função original (que neste caso é chamado * f' *). Isto é geralmente indicado na notação * Lagrange * desta forma:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \end{equation}

Às vezes, você também verá derivadas escritas na notação * Leibniz * desta forma:

\begin{equation}\frac{d}{dx}f(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \end{equation}

*** Nota: *** * Alguns livros usam ** h ** para simbolizar a diferença entre ** x <sub> 0 </sub> ** e ** x <sub> 1 </sub> **, enquanto outros usam ** &Delta; x **. Não faz diferença o valor simbólico que você usa.

#### Forma Alternativa Para Uma Derivada

A fórmula acima mostra a forma generalizada de uma derivada. Você pode usar a função derivada para obter o declive em qualquer ponto dado, por exemplo, para obter o declive no ponto * a * você poderia apenas plugar o valor para * a * na função derivada generalizada:

\begin{equation}f'(\textbf{a}) = \lim_{h \to 0} \frac{f(\textbf{a} + h) - f(\textbf{a})}{h} \end{equation}

Ou você poderia usar a forma alternativa, que é específica para apontar * a *:

\begin{equation}f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a} \end{equation}

Estes são matematicamente equivalentes.

### Encontrando a Derivada Para Um Ponto Específico

É mais fácil entender a diferenciação vendo-a em ação, então vamos usá-la para encontrar a derivada de um ponto específico na função *** f ***.

Aqui está a definição de função *** f ***:

\begin{equation}f(x) = x^{2} + x\end{equation}

Vamos dizer que queremos encontrar *** f '(2) *** (a derivada para *** f *** quando *** x *** é 2); então estamos tentando encontrar a inclinação no ponto mostrado pelo seguinte código:

## Exercício 1 - Onde estiver com a palavra RESOLVER, você deve preencher com o código correspondente.

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Função
def f(x):
    return x**2 + x

# Cria uma matriz de valores x de 0 a 10 para o plot.
x = list(range(0, 11))
y = [f(x_val) for x_val in x]

# Use a função para obter os valores y
# RESOLVER
def f_derivada(x):
    return 2 * x + 1

# Define os pontos
# RESOLVER
x1 = 2
y1 = f(x1)
m = f_derivada(x1)
xMin = x1 - 3
yMin = y1 - (3*m)
xMax = x1 + 3
yMax = y1 + (3*m)


# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot da função
plt.plot(x,y, color='green')

# Plot do ponto
plt.scatter(x1,y1, c='red')
plt.annotate('(x,f(x))',(x1,y1), xytext=(x1-0.5, y1+3))

plt.plot([xMin,xMax],[yMin,yMax], color='magenta')

plt.show()
```

Aqui está nossa fórmula generalizada para encontrar uma derivada em um ponto específico (* a *):

\begin{equation}f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h} \end{equation}

Então vamos começar ligando nosso valor * a * em:

\begin{equation}f'(\textbf{2}) = \lim_{h \to 0} \frac{f(\textbf{2} + h) - f(\textbf{2})}{h} \end{equation}

Sabemos que *** f (x) *** encapsula a equação *** x <sup> 2 </sup> + x ***, então podemos reescrever nossa equação derivada como:

\begin{equation}f'(2) = \lim_{h \to 0} \frac{((2+h)^{2} + 2 + h) - (2^{2} + 2)}{h} \end{equation}

Podemos aplicar a propriedade de distribuição para *** (2 + h) <sup> 2 </sup> *** usando a regra que * (a + b) <sup> 2 </sup> = a <sup> 2 </sup> + b <sup> 2 </sup> + 2ab *:

\begin{equation}f'(2) = \lim_{h \to 0} \frac{(4 + h^{2} + 4h + 2 + h) - (2^{2} + 2)}{h} \end{equation}

Então podemos simplificar 2 <sup> 2 </sup> + 2 (2 <sup> 2 </sup> é 4, mais 2 dá 6):

\begin{equation}f'(2) = \lim_{h \to 0} \frac{(4 + h^{2} + 4h + 2 + h) - 6}{h} \end{equation}

Podemos combinar termos semelhantes no lado esquerdo do numerador para tornar as coisas um pouco mais claras:

\begin{equation}f'(2) = \lim_{h \to 0} \frac{(h^{2} + 5h + 6) - 6}{h} \end{equation}

Que combina ainda mais para se livrar do * 6 *:

\begin{equation}f'(2) = \lim_{h \to 0} \frac{h^{2} + 5h}{h} \end{equation}

E finalmente, podemos simplificar a fração:

\begin{equation}f'(2) = \lim_{h \to 0} h + 5 \end{equation}

Para obter o limite quando * h * está se aproximando de 0, podemos usar a substituição direta para h:

\begin{equation}f'(2) = 0 + 5 \end{equation}

assim:

\begin{equation}f'(2) = 5 \end{equation}

Vamos desenhar uma linha tangente com a inclinação em nosso gráfico para ver se está certo:

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Função
def f(x):
    return x**2 + x

# Cria uma matriz de valores x de 0 a 10 para o plot.
x = list(range(0, 11))

# Use a função para obter os valores y
y = [f(i) for i in x]

# Define os pontos
x1 = 2
y1 = f(x1)

# Especifique a derivada que calculamos acima
m = 5

# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot da função
plt.plot(x,y, color='green')

# Plot do ponto
plt.scatter(x1,y1, c='red')
plt.annotate('(x,f(x))',(x1,y1), xytext=(x1-0.5, y1+3))

# Plot da linha tangente usando a derivada que calculamos
xMin = x1 - 3
yMin = y1 - (3*m)
xMax = x1 + 3
yMax = y1 + (3*m)
plt.plot([xMin,xMax],[yMin,yMax], color='magenta')

plt.show()
```

### Como Encontrar a Derivada Para Qualquer Ponto

Agora vamos juntar tudo e definir uma função que podemos usar para encontrar a derivada para qualquer ponto na função *** f ***:

Aqui está nossa função derivada geral novamente:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \end{equation}

Sabemos que *** f (x) *** encapsula a equação *** x <sup> 2 </sup> + x ***, então podemos reescrever nossa equação derivada como:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{((x+h)^{2} + x + h) - (x^{2} + x)}{h} \end{equation}

Podemos aplicar a propriedade de distribuição para *** (x + h) <sup> 2 </sup> *** usando a regra que * (a + b) <sup> 2 </sup> = a <sup> 2 </sup> + b <sup> 2 </sup> + 2ab *:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{(x^{2} + h^{2} + 2xh + x + h) - (x^{2} + x)}{h} \end{equation}

Então podemos usar a propriedade distributiva para expandir *** - (x <sup> 2 </sup> + x) ***, que é a mesma coisa que * -1 (x <sup> 2 </sup> + x) *, para *** - x <sup> 2 </sup> - x ***:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{x^{2} + h^{2} + 2xh + x + h - x^{2} - x}{h} \end{equation}

Podemos combinar termos semelhantes no numerador para tornar as coisas um pouco mais claras:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{h^{2} + 2xh + h}{h} \end{equation}

E finalmente, podemos simplificar a fração:

\begin{equation}f'(x) = \lim_{h \to 0} 2x + h + 1 \end{equation}

Para obter o limite quando * h * está se aproximando de 0, podemos usar a substituição direta para h:

\begin{equation}f'(x) = 2x + 0 + 1 \end{equation}

assim:

\begin{equation}f'(x) = 2x + 1 \end{equation}

Agora temos uma função para a derivada de *** f ***, que podemos aplicar a qualquer valor * x * para encontrar a inclinação da função em *** f (x ***).

Por exemplo, vamos encontrar a derivada de *** f *** com um valor * x * de 5:

\begin{equation}f'(5) = 2\cdot5 + 1 = 10 + 1 = 11\end{equation}

Vamos usar Python para definir as funções *** f (x) *** e *** f '(x) ***, plotar *** f (5) *** e mostrar a linha tangente para *** f '(5) ***.

## Derivadas de Equações

Nós temos falado sobre derivadas de * funções *, mas é importante lembrar que funções são apenas operações nomeadas que retornam um valor. Podemos aplicar o que sabemos sobre o cálculo de derivadas para qualquer equação, por exemplo:

\begin{equation}\frac{d}{dx}(2x + 6)\end{equation}

Note que geralmente mudamos para a notação * Leibniz * ao encontrar derivadas de equações que não são encapsuladas como funções; mas a abordagem para resolver este exemplo é exatamente a mesma que se tivéssemos uma função hipotética com a definição * 2x + 6 *:

\begin{equation}\frac{d}{dx}(2x + 6) = \lim_{h \to 0} \frac{(2(x+h) + 6) - (2x + 6)}{h} \end{equation}

Após fatorar o * 2 (x + h) * à esquerda e o * - (2x - 6) * à direita, isto é:

\begin{equation}\frac{d}{dx}(2x + 6) = \lim_{h \to 0} \frac{2x + 2h + 6 - 2x - 6}{h} \end{equation}

Podemos simplificar isso para:

\begin{equation}\frac{d}{dx}(2x + 6) = \lim_{h \to 0} \frac{2h}{h} \end{equation}

Agora podemos fatorar * h * completamente, então em qualquer ponto:

\begin{equation}\frac{d}{dx}(2x + 6) = 2 \end{equation}

Se você executar o código Python abaixo para traçar a linha criada pela equação, verá que ela realmente tem uma inclinação constante de 2.

## Regras e Operações com Derivadas

Ao trabalhar com derivadas, existem algumas regras ou atalhos que você pode aplicar para facilitar sua vida.

### Regras Básicas de Derivadas

Vamos começar com algumas regras básicas que é útil saber.

- Se * f (x) * = * C * (onde * C * é uma constante), então * f '(x) * = 0

    Isso faz sentido se você pensar nisso por um segundo. Não importa o valor que você usa para * x *, a função retorna o mesmo valor constante; então o gráfico da função será uma linha horizontal. Não há taxa de variação em uma linha horiziontal, então sua inclinação é 0 em todos os pontos. Isto é verdade para qualquer constante, incluindo constantes simbólicas como * &pi; * (pi).
    
    Então, por exemplo:
    
\begin{equation}f(x) = 6 \;\; \therefore \;\; f'(x) = 0 \end{equation}

    Ou:

\begin{equation}f(x) = \pi \;\; \therefore \;\; f'(x) = 0 \end{equation}
    
- Se * f (x) * = * Cg (x) *, então * f '(x) * = * Cg' (x) *

    Esta regra nos diz que se uma função é igual a uma segunda função multiplicada por uma constante, então a derivada da primeira função será igual à derivada da segunda função multiplicada pela mesma constante. Por exemplo:
    
\begin{equation}f(x) = 2g(x) \;\; \therefore \;\; f'(x) = 2g'(x) \end{equation}

- Se * f (x) * = * g (x) * + * h (x) *, então * f '(x) * = * g' (x) * + * h '(x) *

    Em outras palavras, se uma função é a soma de duas outras funções, então a derivada da primeira função é a soma das derivadas das outras duas funções. Por exemplo:
    
\begin{equation}f(x) = g(x) + h(x) \;\; \therefore \;\; f'(x) = g'(x) + h'(x) \end{equation}

    Claro, isso também se aplica à subtração:
    
\begin{equation}f(x) = k(x) - l(x) \;\; \therefore \;\; f'(x) = k'(x) - l'(x) \end{equation}

Como discutido anteriormente, as funções são apenas equações encapsuladas como uma entidade nomeada que retorna um valor e as regras podem ser aplicadas a qualquer equação. Por exemplo:

\begin{equation}\frac{d}{dx}(2x + 6) = \frac{d}{dx} 2x +  \frac{d}{dx} 6\end{equation}

Assim, podemos aproveitar as regras para facilitar o cálculo:

\begin{equation}\frac{d}{dx}(2x) = \lim_{h \to 0} \frac{2(x+h) - 2x}{h} \end{equation}

Após fatorar o * 2 (x + h) * à esquerda, isto é:

\begin{equation}\frac{d}{dx}(2x) = \lim_{h \to 0} \frac{2x + 2h - 2x}{h} \end{equation}

Podemos simplificar isso para:

\begin{equation}\frac{d}{dx}(2x) = \lim_{h \to 0} \frac{2h}{h} \end{equation}

O que nos dá:

\begin{equation}\frac{d}{dx}(2x) = 2 \end{equation}

Agora podemos voltar nossa atenção para a derivada da constante 6 em relação a * x *, e sabemos que a derivada de uma constante é sempre 0, então:

\begin{equation}\frac{d}{dx}(6) = 0\end{equation}

Adicionamos as duas derivadas que calculamos:

\begin{equation}\frac{d}{dx}(2x + 6) = 2 + 0\end{equation}

O que nos dá o nosso resultado:

\begin{equation}\frac{d}{dx}(2x + 6) = 2\end{equation}

### A Regra da Potência

A * regra da potência * é um dos atalhos mais úteis no mundo do cálculo diferencial. Pode ser dito assim:

\begin{equation}f(x) = x^{n} \;\; \therefore \;\; f'(x) = nx^{n-1}\end{equation}

Então, se a nossa função para * x * retorna * x * para a potência de alguma constante (que vamos chamar de * n *), então a derivada da função para * x * é * n * vezes * x * para o poder de * n * - 1.

Provavelmente é útil examinar alguns exemplos para ver como isso funciona:

\begin{equation}f(x) = x^{3} \;\; \therefore \;\; f'(x) = 3x^{2}\end{equation}

\begin{equation}f(x) = x^{-2} \;\; \therefore \;\; f'(x) = -2x^{-3}\end{equation}

\begin{equation}f(x) = x^{2} \;\; \therefore \;\; f'(x) = 2x\end{equation}

Em cada um desses exemplos, o exponencial de * x * na definição da função torna-se o coeficiente para * x * na definição de derivada, com o exponencial sendo decrementado por 1.

Aqui está um exemplo trabalhado para encontrar a derivada da seguinte função:

\begin{equation}f(x) = x^{2}\end{equation}

Então começamos com a função derivada geral:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \end{equation}

Nós podemos conectar nossa definição para * f *:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{(x + h)^{2} - x^{2}}{h} \end{equation}

Agora podemos fatorar o binômio quadrado perfeito à esquerda:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{x^{2} + h^{2} + 2xh - x^{2}}{h} \end{equation}

Os termos x <sup> 2 </sup> cancelam um ao outro, então chegamos a:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{h^{2} + 2xh}{h} \end{equation}

O que simplifica para:

\begin{equation}f'(x) = \lim_{h \to 0} h + 2x \end{equation}

Com * h * se aproximando de 0, isto é:

\begin{equation}f'(x) = 0 + 2x \end{equation}

Então nossa resposta é:

\begin{equation}f'(x) = 2x \end{equation}

Note que poderíamos ter alcançado o mesmo resultado simplesmente aplicando a regra da potência e transformando x <sup> 2 </sup> em 2x <sup> 1 </sup> (que é o mesmo que 2x).

### A Regra do Produto

A * regra do produto *  pode ser declarada como:

\begin{equation}\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x) \end{equation}

Isso significa que a derivada de * f (x) * multiplicada por * g (x) * é igual à derivada de * f (x) * multiplicada pelo valor de * g (x) * adicionado ao valor de * g (x) * * f (x) * multiplicado pela derivada de * g (x *).

Vamos ver um exemplo baseado nas duas funções a seguir:

\begin{equation}f(x) = 2x^{2} \end{equation}

\begin{equation}g(x) = x + 1 \end{equation}

Vamos começar calculando a derivada de * f (x) *:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{2(x + h)^{2} - 2x^{2}}{h} \end{equation}

Isso fatora para:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{2x^{2} + 2h^{2} + 4xh - 2x^{2}}{h} \end{equation}

Que quando cancelamos 2x <sup> 2 </sup> e -2x <sup> 2 </sup> é:

\begin{equation}f'(x) = \lim_{h \to 0} \frac{2h^{2} + 4xh}{h} \end{equation}

O que simplifica para:

\begin{equation}f'(x) = \lim_{h \to 0} 2h + 4x \end{equation}

Com * h * se aproximando de 0, isto é:

\begin{equation}f'(x) = 4x \end{equation}

Agora vamos olhar para * g '(x) *:

\begin{equation}g'(x) = \lim_{h \to 0} \frac{(x + h) + 1 - (x + 1)}{h} \end{equation}

Podemos apenas remover os colchetes à esquerda e fatorar o * - (x + 1) * à direita:

\begin{equation}g'(x) = \lim_{h \to 0} \frac{x + h + 1 - x - 1}{h} \end{equation}

Qual pode ser simplificado até:

\begin{equation}g'(x) = \lim_{h \to 0} \frac{h}{h} \end{equation}

Habilitando-nos a fatorar * h * completamente para dar uma derivada constante de * 1 *:

\begin{equation}g'(x) = 1 \end{equation}

Então, agora podemos calcular a derivada para o produto dessas funções, ligando as funções e as derivadas que calculamos para elas na equação da regra do produto:

\begin{equation}\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x) \end{equation}

Assim:

\begin{equation}\frac{d}{dx}[f(x)g(x)] = (4x \cdot (x + 1)) + (2x^{2} \cdot 1) \end{equation}

Que pode ser simplificado para:

\begin{equation}\frac{d}{dx}[f(x)g(x)] = (4x^{2} + 4x) + 2x^{2} \end{equation}

Que pode ser ainda mais simplificado para:

\begin{equation}\frac{d}{dx}[f(x)g(x)] = 6x^{2} + 4x \end{equation}

### A Regra do Quociente

A * regra do quociente * aplica-se a funções que são definidas como um quociente de uma expressão dividida por outra, por exemplo:;

\begin{equation}r(x) = \frac{s(x)}{t(x)} \end{equation}

Nesta situação, você pode aplicar a seguinte regra de quociente para encontrar a derivada de * r (x) *:

\begin{equation}r'(x) = \frac{s'(x)t(x) - s(x)t'(x)}{[t(x)]^{2}} \end{equation}

Aqui estão as nossas definições para * s (x) * e * t (x) *:

\begin{equation}s(x) = 3x^{2} \end{equation}

\begin{equation}t(x) = 2x\end{equation}

Vamos começar com * s'(x) *:

\begin{equation}s'(x) = \lim_{h \to 0} \frac{3(x + h)^{2} - 3x^{2}}{h} \end{equation}

Isso fatora para:

\begin{equation}s'(x) = \lim_{h \to 0} \frac{3x^{2} + 3h^{2} + 6xh - 3x^{2}}{h} \end{equation}

Que quando cancelamos 3x <sup> 2 </sup> e -3x <sup> 2 </sup> é:

\begin{equation}s'(x) = \lim_{h \to 0} \frac{3h^{2} + 6xh}{h} \end{equation}

O que simplifica para:

\begin{equation}s'(x) = \lim_{h \to 0} 3h + 6x \end{equation}

Com * h * se aproximando de 0, isto é:

\begin{equation}s'(x) = 6x \end{equation}

Agora vamos olhar para * t'(x) *:

\begin{equation}t'(x) = \lim_{h \to 0} \frac{2(x + h) - 2x}{h} \end{equation}

Podemos apenas fatorar o * 2 (x + h) * à esquerda:

\begin{equation}t'(x) = \lim_{h \to 0} \frac{2x + 2h - 2x}{h} \end{equation}

Qual pode ser simplificado até:

\begin{equation}t'(x) = \lim_{h \to 0} \frac{2h}{h} \end{equation}

Habilitando-nos a fatorar * h * completamente para dar uma derivada constante de * 2 *:

\begin{equation}t'(x) = 2 \end{equation}

Então, agora podemos calcular a derivada para o quociente dessas funções, ligando as definições de função e as derivadas que calculamos para elas na equação da regra do quociente:

\begin{equation}r'(x) = \frac{(6x \cdot 2x) - (3x^{2} \cdot 2)}{[2x]^{2}} \end{equation}

Podemos fatorar os termos do numerador assim:

\begin{equation}r'(x) = \frac{12x^{2} - 6x^{2}}{[2x]^{2}} \end{equation}

Que então pode ser combinado:

\begin{equation}r'(x) = \frac{6x^{2}}{[2x]^{2}} \end{equation}

O denominador é [2x] <sup> 2 </sup> (note que isso é diferente de 2x <sup> 2 </sup>. [2x] <sup> 2 </sup> é 2x & bull; 2x, enquanto 2x <sup> 2 </sup> é 2 &bull; x <sup> 2 </sup>):

\begin{equation}r'(x) = \frac{6x^{2}}{4x^{2}} \end{equation}

O que simplifica para:

\begin{equation}r'(x) = 1\frac{1}{2} \end{equation}

Então a derivada de * r (x) * é 1.5.

### A Regra da Cadeia

A * regra da cadeia * aproveita o fato de que as equações podem ser encapsuladas como funções, e como as funções contêm equações, é possível aninhar uma função dentro da outra.

Por exemplo, considere a seguinte função:

\begin{equation}u(x) = 2x^{2} \end{equation}

Poderíamos ver a definição de * u(x) * como um composto de duas funções; uma função * inner * que calcula x <sup> 2 </sup> e uma função * outer * que multiplica o resultado da função interna por 2.

\begin{equation}u(x) = \widehat{\color{blue}2\color{blue}(\underline{\color{red}x^{\color{red}2}}\color{blue})} \end{equation}

Para tornar as coisas mais simples, podemos nomear essas funções internas e externas:

\begin{equation}i(x) = x^{2} \end{equation}

\begin{equation}o(x) = 2x \end{equation}

Note que * x * indica a entrada para cada função. A função * i * recebe sua entrada e eleva ao quadrado, e a função * o * pega sua entrada e a multiplica por 2. Quando usamos isso como uma função composta, a entrada do valor * x * na função externa será a saída da função interna.

Vamos dar uma olhada em como podemos aplicar essas funções para voltar à nossa função * u * original:

\begin{equation}u(x) = o(i(x)) \end{equation}

Então, primeiro precisamos encontrar a saída da função interna * i * para que possamos usar como o valor de entrada para a função externa * o *. Bem, isso é fácil, nós sabemos a definição de * i * (o quadrado da entrada), então podemos simplesmente ligá-lo:

\begin{equation}u(x) = o(x^{2}) \end{equation}

Também conhecemos a definição para a função externa * o * (multiplique a entrada por 2), para que possamos aplicar isso à entrada:

\begin{equation}u(x) = 2x^{2} \end{equation}

OK, agora sabemos como formar uma função composta. A * regra da cadeia * pode ser declarada assim:

\begin{equation}\frac{d}{dx}[o(i(x))] = o'(i(x)) \cdot i'(x)\end{equation}

Tudo bem, vamos começar ligando a saída da função interna * i (x) * em:

\begin{equation}\frac{d}{dx}[o(i(x))] = o'(x^{2}) \cdot i'(x)\end{equation}

Agora vamos usar isso para calcular a derivada de * o *, substituindo cada * x * na equação com a saída da função * i * (* x <sup> 2 </sup> *):

\begin{equation}o'(x) = \lim_{h \to 0} \frac{2(x^{2} + h) - 2x^{2}}{h} \end{equation}

Isso fatora para:

\begin{equation}o'(x) = \lim_{h \to 0} \frac{2x^{2} + 2h - 2x^{2}}{h} \end{equation}

Que quando cancelamos 2x <sup> 2 </sup> e -2x <sup> 2 </sup> é:

\begin{equation}o'(x) = \lim_{h \to 0} \frac{2h}{h} \end{equation}

O que simplifica para:

\begin{equation}o'(x) = 2 \end{equation}

Agora podemos calcular * i '(x) *. Sabemos que a definição de * i (x) * é x <sup> 2 </sup>, e podemos usar a regra da potência para determinar que * i '(x) * é, portanto, 2x.

Então nossa equação neste ponto é:

\begin{equation}\frac{d}{dx}[o(i(x))] = 2 \cdot 2x\end{equation}

A qual é:

\begin{equation}\frac{d}{dx}[o(i(x))] = 4x\end{equation}

Comumente, a regra da cadeia é declarada usando uma notação ligeiramente diferente com a qual você pode achar mais fácil trabalhar. Nesse caso, podemos pegar nossa equação:

\begin{equation}\frac{d}{dx}[o(i(x))] = o'(i(x)) \cdot i'(x)\end{equation}

e reescrevê-lo como

\begin{equation}\frac{du}{dx} = \frac{do}{di}\frac{di}{dx}\end{equation}

Podemos então completar os cálculos da seguinte forma:

\begin{equation}\frac{du}{dx} = 2 \cdot 2x = 4x\end{equation}

## Direção da Função em um Ponto

Considere a seguinte função, que representa a trajetória de uma bola que foi chutada em um campo de futebol:

\begin{equation}k(x) = -10x^{2} + 100x + 3 \end{equation}

Execute o código Python abaixo para representar graficamente essa função e ver a trajetória da bola ao longo de um período de 10 segundos.

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Cria a Função
def k(x):
    return -10*(x**2) + (100*x)  + 3

# Cria uma matriz de valores x para o plot
x = list(range(0, 11))

# Use a função para obter os valores y
y = [k(i) for i in x]

# Define o gráfico
plt.xlabel('x (tempo em segundos)')
plt.ylabel('k(x) (altura em centímetros)')
plt.xticks(range(0,15, 1))
plt.yticks(range(-200, 500, 20))
plt.grid()

# Plot 
plt.plot(x,y, color='green')

plt.show()
```

Ao olhar para o gráfico desta função, você pode ver que ela descreve uma parábola na qual a bola subiu em altura antes de cair de volta ao chão. No gráfico, é bastante fácil ver quando a bola estava subindo e quando estava caindo.

Naturalmente, também podemos usar derivadas para determinar a inclinação da função em qualquer ponto. Podemos aplicar algumas das regras que discutimos anteriormente para determinar a função derivada:

- Podemos somar as derivadas dos termos individuais (*** - 10x <sup> 2 </sup> ***, *** 100x *** e *** 3 ***) para encontrar a derivada da função inteira.
- A * regra da potência * nos diz que a derivada de *** - 10x <sup> 2 </sup> *** é *** - 10 &bull; 2x ***, que é *** - 20x ***.
- A * regra da potência * também nos diz que a derivada de *** 100x *** é *** 100 ***.
- A derivada de qualquer constante, como *** 3 *** é *** 0 ***.

Assim:

\begin{equation}k'(x) = -20x + 100 + 0 \end{equation}

O que obviamente simplifica para:

\begin{equation}k'(x) = -20x + 100 \end{equation}

Agora podemos usar essa função derivada para encontrar a inclinação para qualquer valor de *** x ***.

Execute a célula abaixo para ver um gráfico da função e sua função derivada:

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Funções
def k(x):
    return -10*(x**2) + (100*x)  + 3

def kd(x):
    return -20*x + 100

# Cria uma matriz de valores x para o plot
x = list(range(0, 11))

# Use a função para obter os valores y
y = [k(i) for i in x]

# Use a função derivada para obter as derivadas
yd = [kd(i) for i in x]

# Define o gráfico
plt.xlabel('x (tempo em segundos)')
plt.ylabel('k(x) (altura em centímetros)')
plt.xticks(range(0,15, 1))
plt.yticks(range(-200, 500, 20))
plt.grid()

# Plot da função
plt.plot(x,y, color='green')

# Plot da derivada
plt.plot(x,yd, color='purple')

plt.show()
```

Observe atentamente a linha roxa que representa a função derivada e observe que ela é um valor decrescente constante - em outras palavras, a inclinação da função está reduzindo linearmente à medida que x aumenta. Observe também que o ponto em que a linha derivada cruza 0 no eixo y é também o ponto em que o valor da função pára de aumentar e começa a diminuir. Quando a inclinação tem um valor positivo, a função está aumentando; e quando a inclinação tem um valor negativo, a função está diminuindo.

O fato de a linha derivada cruzar 0 no ponto mais alto da função faz sentido se você pensar nela logicamente. Se você desenhasse a linha tangente representando a inclinação em cada ponto, estaria girando no sentido horário ao longo do gráfico, apontando inicialmente para cima e para a direita à medida que a bola se ergue e girando até apontar para baixo e para a direita quando a bola cai. ## Pontos Críticos

Pelo que vimos até agora, parece que existe uma relação entre uma função que atinge um valor extremo (máximo ou mínimo) e um valor derivativo de 0. Isso faz sentido intuitivo; a derivada representa a inclinação da linha, portanto, quando uma função muda de uma inclinação negativa para uma inclinação positiva, ou vice-versa, a derivada deve passar por 0.

No entanto, você precisa ter cuidado para não assumir que apenas porque a derivada é 0 em um determinado ponto, esse ponto representa o mínimo ou máximo da função. Por exemplo, considere a seguinte função:

\begin{equation}v(x) = x^{3} - 2x + 100 \end{equation}

Execute o seguinte código Python para visualizar esta função e sua função derivada correspondente.

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Funções
def v(x):
    return (x**3) - (2*x) + 100

def vd(x):
    return 3*(x**2) - 2

# Cria uma matriz de valores x para o plot
x = list(range(-10, 11))

# Use a função para obter os valores y
y = [v(i) for i in x]

# Use a função derivada para obter os valores das derivadas
yd = [vd(i) for i in x]

# Define o gráfico
plt.xlabel('x')
plt.ylabel('v(x)')
plt.xticks(range(-10,15, 1))
plt.yticks(range(-1000, 2000, 100))
plt.grid()

# Plot ds função
plt.plot(x,y, color='green')

# Plot da derivada
plt.plot(x,yd, color='purple')

plt.show()
```

Observe que, nesse caso, a linha da função derivada roxa passa por 0 à medida que a linha da função verde passa de um declive * côncavo para baixo * (um declive que está diminuindo) para um declive * côncavo para cima * (um declive que aumenta). O declive se estabiliza em 0, formando uma "reta" antes de começar a aumentar.

O que podemos aprender com isso é que coisas interessantes parecem acontecer com a função quando a derivada é 0. Chamamos pontos onde a derivada cruza 0 * pontos críticos *, porque eles indicam que a função está mudando de direção. Quando uma função muda de direção de positiva para negativa, ela forma um pico (ou um * máximo local *), quando a função muda de direção de negativa para positiva forma um vale (ou * mínimo local *) e quando mantém a mesma direção geral, mas altera a concavidade da inclinação, cria um * ponto de inflexão * (saddle).

## Encontrando Mínima e Máxima

Um uso comum do cálculo é encontrar pontos mínimos e máximos em uma função. Por exemplo, poderíamos querer saber quantos segundos levou para que a bola chutada no jogo de futebol alcançasse sua altura máxima, ou quanto tempo demorou para que nosso fertilizante fosse efetivo em reverter o declínio do crescimento de flores.

Vimos que quando uma função muda de direção para criar um pico máximo ou mínimo, a derivada da função é 0, então um passo em direção a encontrar esses pontos extremos pode ser simplesmente encontrar todos os pontos na função onde a função derivada é 0. Por exemplo, aqui está nossa função para o chute da bola:

\begin{equation}k(x) = -10x^{2} + 100x + 3 \end{equation}

A partir disso, calculamos a função para a derivada como:

\begin{equation}k'(x) = -20x + 100 \end{equation}

Podemos então resolver a equação derivada para um valor f '(x) de 0:

\begin{equation}-20x + 100 = 0 \end{equation}

Podemos remover a constante subtraindo 100 a ambos os lados:

\begin{equation}-20x = -100 \end{equation}

Multiplicando ambos os lados por -1, elimina os valores negativos (isto não é estritamente necessário, mas torna a equação um pouco menos confusa)

\begin{equation}20x = 100 \end{equation}

Assim:

\begin{equation}x = 5 \end{equation}

Então sabemos que a derivada será 0 quando * x * é 5, mas isso é um mínimo, um máximo ou nenhum dos dois? Poderia ser apenas um ponto de inflexão, ou toda a função poderia ser um valor constante com uma inclinação de 0. Sem olhar para o gráfico, é difícil dizer.

## Derivada de Segunda Ordem

A solução para o nosso problema é encontrar a derivada da derivada! Até agora, encontramos a derivada de uma função e a indicamos como *** f'(x) ***. Tecnicamente, isso é conhecido como o derivado * primo *; e descreve a inclinação da função. Como a função derivada é em si uma função, podemos encontrar sua derivada, que chamamos de derivada de segunda ordem (ou, às vezes, apenas segunda). Isso é indicado assim: *** f '' (x) ***.

\begin{equation}k''(x) = -20 \end{equation}

Agora, sem desenhar o gráfico, podemos ver que a segunda derivada tem um valor constante; então sabemos que a inclinação da derivada principal é linear; e porque é um valor negativo, sabemos que está diminuindo. Assim, quando a derivada principal cruza 0, sabemos que a inclinação da função está diminuindo linearmente; então o ponto em * x = 0 * deve ser um ponto máximo.

## Otimização

A capacidade de usar derivadas para encontrar os mínimos e máximos de uma função a torna uma ferramenta útil para cenários em que você precisa otimizar uma função para uma variável específica.

### Definindo Funções para Otimização

Por exemplo, suponha que você tenha decidido criar um serviço de vídeo on-line baseado em um modelo de assinatura. Você planeja cobrar uma taxa de assinatura mensal e deseja obter a maior receita possível. O problema é que os clientes são sensíveis ao preço, por isso, se você definir a taxa mensal muito alta, impedirá que alguns clientes se inscrevam. Por outro lado, se você definir a taxa muito baixa, poderá obter mais clientes, mas ao custo da receita reduzida.

O que você precisa é de algum tipo de função que lhe dirá quantas assinaturas você espera obter com base em uma determinada taxa. Então você fez alguma pesquisa e encontrou uma fórmula para indicar que o volume de assinatura esperado (em milhares) pode ser calculado como 5 vezes a taxa mensal subtraída de 100; ou expresso como uma função:

\begin{equation}s(x) = -5x + 100\end{equation}

O que você realmente deseja otimizar é a receita mensal, que é simplesmente o número de assinantes multiplicado pela taxa:

\begin{equation}r(x) = s(x) \cdot x\end{equation}

Nós podemos combinar *** s (x) *** em *** r (x) *** assim:

\begin{equation}r(x) = -5x^{2} + 100x\end{equation}

### Encontrando a Derivada Principal

A função *** r(x) *** retornará a receita mensal esperada (em milhares) para qualquer taxa proposta (* x *). O que precisamos fazer agora é encontrar a taxa que gera a receita máxima. Felizmente, podemos usar uma derivada para fazer isso.

Primeiro, precisamos determinar a derivada principal de *** r(x) ***, e podemos fazer isso facilmente usando a regra da potência:

\begin{equation}r'(x) = 2 \cdot -5x + 100\end{equation}

A qual é:

\begin{equation}r'(x) = -10x + 100\end{equation}

### Encontrando Pontos Críticos

Agora precisamos encontrar quaisquer pontos críticos onde a derivada é 0, pois isso pode indicar um máximo:

\begin{equation}-10x + 100 = 0\end{equation}

Vamos isolar o termo * x *:

\begin{equation}-10x = -100\end{equation}

Ambos os lados são negativos, então podemos multiplicar ambos por -1 para torná-los positivos sem afetar a equação:

\begin{equation}10x = 100\end{equation}

Agora podemos dividir ambos os lados por 10 para isolar * x *:

\begin{equation}x = \frac{100}{10}\end{equation}

Assim:

\begin{equation}x = 10\end{equation}

#### Verificando se há um Máximo

Agora sabemos que com um valor * x * de ** 10 **, a derivada é 0; ou dito de outra forma, quando a taxa é 10, a inclinação indicando a mudança no volume de assinatura é plana. Isso poderia ser um ponto em que a mudança no volume de assinaturas atingiu o pico (em outras palavras, um máximo); mas também pode ser um ponto mínimo ou apenas um ponto de inflexão em que a taxa de mudança muda de negativa para positiva.

Para ter certeza, podemos verificar a derivada de segunda ordem. Podemos calcular isso aplicando a regra da potência à derivada principal:

\begin{equation}r''(x) = -10\end{equation}

Observe que a segunda derivada é uma constante com um valor negativo. Será o mesmo para qualquer ponto, incluindo nosso ponto crítico em * x = 10 *:

\begin{equation}r''(10) = -10\end{equation}

Um valor negativo para a segunda derivada nos diz que a inclinação derivativa está se movendo em uma direção negativa no ponto em que é 0, portanto, o valor da função deve estar no máximo.

Em outras palavras, a taxa mensal ideal para o nosso serviço de vídeo on-line é 10 - isso gerará a receita mensal máxima.

Execute o código abaixo para mostrar a função *** r(x) *** como um gráfico e verifique se o ponto máximo está em x = 10.

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Função s
def s(x):
    return (-5*x) + 100

# Função r
def r(x):
    return s(x) * x

# Cria uma matriz de valores x para o plot
x = list(range(0, 21))

# Use a função para obter os valores y
y = [r(i) for i in x]

# Define o gráfico
plt.xlabel('x (taxa mensal)')
plt.ylabel('r(x) (faturamento em R$,000)')
plt.xticks(range(0,22, 1))
plt.yticks(range(0, 600, 50))
plt.grid()

# Plot da função
plt.plot(x,y, color='green')

plt.show()
```

# Cálculo Integral

Integrais são os inversos de derivadas. Mais importante, o uso da integração fornece uma maneira de calcular a área sob a curva de praticamente qualquer função. Existem muitas aplicações para integração. Por exemplo, se você precisar calcular uma probabilidade de alguma ocorrência entre limites (que discutiremos mais adiante neste curso), você usará uma integral.

Vamos começar com uma função simples:

$$ f (x) = x $$

Podemos traçar essa função como uma linha. Execute o código abaixo para traçar a função para o intervalo de 0 a 10.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# Função
def f(x):
    return x

# Cria uma matriz de valores x de 0 a 10
x = range(0, 11)

# Obtém os valores y correspondentes da função
y = [f(a) for a in x]

# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot x e f(x)
plt.plot(x,y, color='purple')

plt.show()
```

## Executando Integração

O * integral * de uma função é a área sob ela - nesse caso, a área sob a linha diagonal roxa até o eixo x.

Então, como você encontra a integral de uma função? bem, para nossa função simples $ f(x) = x $, a fórmula para uma integral é escrita da seguinte forma:

$$\int f(x)\;dx$$

O símbolo $\int$ mostra que esta fórmula é uma integral. O $dx$ indica que a integração é em relação à variável $x$. Observe que, como $f(x) = x$, também poderíamos escrever essa fórmula integral como $\int x\;dx$

Então, qual é a integral de $x\ dx$? Para responder a essa pergunta, precisamos da * antiderivada * de *** f *** - em outras palavras, precisamos encontrar uma função que tenha uma derivada que corresponda à saída de $f$, que é apenas $x$. Usando a regra da potência no sentido inverso, uma função que tenha a derivada $x$ seria $\frac{1}{2} x^2$

Então, a fórmula da integral para $f$ em relação a $x$ pode ser escrita como:

$$\int f(x)\;dx = \frac{1}{2} x^2$$

## Integração entre limites

Agora que temos a fórmula integral, podemos usá-la para encontrar a integral entre pontos iniciais e finais específicos. Vamos supor que queremos encontrar a área sob a função entre os valores $x$ 0 e 2. Em outras palavras, o * integral * de $f$ para o intervalo de 0 a 2 em relação a $x$.

Execute o seguinte código para reformular a função e mostrar a área em que estamos interessados:

```python
# Define o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot x e f(x)
plt.plot(x,y, color='purple')

# Mostra a área para a integral
section = np.arange(0, 2, 1/20)
plt.fill_between(section,f(section), color='orange')

plt.show()
```

Chamamos o ponto inicial e final os ** limites ** da integral. O limite inferior é colocado como um subscrito do sinal integral. O limite superior é colocado como um sobrescrito do sinal integral. Usando essa notação, a integral de $f(x)$ de $0$ para $2$ é escrita da seguinte maneira:

$$\int_0^2 f(x)\;dx$$

A integral é avaliada subtraindo-se o valor do integrando no limite inferior do integrante no limite superior; e como sabemos a fórmula baseada em nossa função antiderivada, a integral pode ser avaliada da seguinte maneira.

$$\int_0^2 f(x)\;dx\\ = \frac{1}{2} x^2\ \big|_0^2\\
= \frac{1}{2} 2^2 - \frac{1}{2} 0^2\\
= \frac{4}{2} - \frac{0}{2} x^2\\ = 2$$

Vamos olhar para essa função e a área em questão:

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Polygon
%matplotlib inline

# Função
def g(x):
    return 3 * x**2 + 2 * x + 1

# Cria uma matriz de valores x de 0 a 10
x = range(0, 11)

# Obtém os valores y correspondentes da função
y = [g(a) for a in x]

# Define o gráfico
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot x e g(x)
plt.plot(x,y, color='purple')

# Desenha a região sombreada
ix = np.linspace(0, 3)
iy = g(ix)
verts = [(0, 0)] + list(zip(ix, iy)) + [(3, 0)]
poly = Polygon(verts, facecolor='orange')
ax.add_patch(poly)

plt.show()
```

Podemos avaliar esta integral como antes, desta vez usando a função:

$$\int_0^3 3x^2 + 2x + 1\;dx$$ 

Podemos calcular a antiderivada de $3x^2 + 2x + 1\;dx$ como $\frac{3}{3} x^3 + \frac{2}{2} x^2 + x$, portanto:

= \frac{3}{3} 3^3 + \frac{2}{2} 3^2 + 3 - \frac{3}{3} 0^3 - \frac{2}{2} 0^2 + 0\\
= 27 + 9 + 3 + 0 + 0 + 0\\
